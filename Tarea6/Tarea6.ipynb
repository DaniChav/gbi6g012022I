## Descendencia (500 generaciones)

```python
# Se generan 500 descendencias a partir de una generación inicial predeterminada con una población de 20 individuos (N)
# y una probabilidad de incidencia del alelo dominante A de 0.55 (p)
N=500
for i in range(0, 500):
    new_pop = pm.reproduce_population(pm.build_population(N=20, p=0.55))
    print(i, new_pop)
```

```python
new_pop = pm.reproduce_population(my_pop)
N=500
for i in range(0, 500):
    fnew_pop = pm.compute_frequencies(pm.build_population(N=20, p=0.55))
    print(i, fnew_pop)
```

```python
# Proporciones alélicas (AA, Aa, aA, aa) de las 500 generaciones
print("AA", (500-fnew_pop["AA"])/500)
print("Aa", (500-fnew_pop["Aa"])/500)
print("aA", (500-fnew_pop["aA"])/500)
print("aa", (500-fnew_pop["aa"])/500)
```
